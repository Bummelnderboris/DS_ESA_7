```python
#import our stuff and have a look at the dataframe 
import pandas as pd
import matplotlib as plt
import numpy as np
#reminder: NumPy arrays have one dtype for the entire array, while pandas DataFrames have one dtype per column.

df = pd.read_csv ("datasets_22520_28884_Grid_Disruption_00_14_standardized_Grid_Disruption_00_14_standardized.csv")

#drop duplicates if there are some
df = df.drop_duplicates()

#have a look at the first few rows
df.head(20)
```

So: Where are errors here? How would you clean this file?

It is observable that there are aswell Unknown as NaN Values given. Because of the consistence of this pattern it is assumed we need this disticntion, so changing the dtype to a date for columns like "Date Event Began" won't be a usefull strategy. This circumstance makes it a bit difficult to plot or analyse data because of non consistent data types in one column, so we have them stored as objects. Assuming we need to keep the information if a value is unknown, there are three different possibilities. We could create a new column for the values with matching dtype. We could keep the data general and def a query , for example with try and except. We could asign nonsense values within the matching type to Unknown values so it's obious these are made up values, but you'd need to pass on the information, that these values are pointing out unknonwn values.

The first possibility would create a lot of redundant data in the same data frame and the third one might be inconstitent on longterm effects so I suggest to use the second strategy and handle unknown values within the query, or create a whole new date frame with cleaned values, what we will do as a second attempt.

We also need to know how many "Unknown" objects are listed in the dataset.

```python
#before doing so lets have a quick look at the data types
df.dtypes

#as expected we have a lot of objects because of the different contents in one column
```

```python
#Now lets distinguish between empty and unknown vales

#empty
all_v = np.product(df.shape)
miss_v = df.isnull().sum()


per_v = miss_v/all_v*100
print("List of empty values in %:")
per_v
```

```python
#unknown
unk_v = df == "Unknown"

un_co_v = unk_v.sum()

print("List of unkown values in %:")
un_co_v / all_v*100
```

```python
#To avoid missunderstandings in future we will clear all missing seels and replace them with a blankspace.
df = df.fillna("")
df.head(20)
```

```python
#doublecheck if we filled all missing values
miss_v = df.isnull().sum()

per_v = miss_v/all_v*100
print("List of empty values in %:")
per_v
```

```python
#Now we got a dataframe with the disticntion, if a value is unknown or not given.
df.head(20)
```

```python
df1 = pd.read_csv ("datasets_22520_28884_Grid_Disruption_00_14_standardized_Grid_Disruption_00_14_standardized.csv")
df1
```

```python
#SHIT.. For the first attempt we had a look at our data frame by viewing at the first 20 rows (.head(20))
# now we see the date time column contains different versions of writing for AM/PM and a.m./p.m.
# let's keep that in mind

#okay lets try to use the pandas build in function .convert_dtypes() to convert into propper types.
df1 = df1.convert_dtypes()
df1.dtypes
```

```python
df1 = df1.convert_dtypes()
df1.dtypes
```

```python
#Okay our problem is not solved, so we gonna make it manually
```

```python
#the pandas function .conver_dtypes() didn't help as as requested bu turned the objects into strings,
#so we can use the str.replace() function to change a.m. only in the string to AM (same for pm)

df1["Time Event Began"] = df1["Time Event Began"].str.replace("a.m.", "AM")
df1["Time Event Began"] = df1["Time Event Began"].str.replace("p.m.", "PM")
df1
```

```python
#Perfect now we have it uniform
```
